```python
import tensorflow as tf
import numpy as np
import cv2
labels = ["Acne Papule",
  "Blackheads",
  "Fungal Acne",
  "Milia",
  "Normal",
  "Perioral Dermatitis",
  "Sun Spots",
  "Whiteheads"]
class ImageClassifier:

    PROBABILITY_MEAN = 0
    PROBABILITY_STD = 255.0
    IMAGE_STD = 1.0
    IMAGE_MEAN = 0.0
    MAX_SIZE = 8

    def __init__(self, model_path, labels_path):
        self.model = tf.lite.Interpreter(model_path=model_path)
        self.model.allocate_tensors()
        self.labels = labels
        self.input_details = self.model.get_input_details()
        self.output_details = self.model.get_output_details()
        self.image_resize_x = self.input_details[0]['shape'][1]
        self.image_resize_y = self.input_details[0]['shape'][2]
        self.input_image_buffer = np.zeros((1, self.image_resize_x, self.image_resize_y, 3), dtype=np.uint8)

    def recognize_image(self, image_path):
        image = cv2.imread(image_path)
        input_image = self.load_image(image)
        self.input_image_buffer[0] = input_image
        self.model.set_tensor(self.input_details[0]['index'], self.input_image_buffer)
        self.model.invoke()
        output_data = self.model.get_tensor(self.output_details[0]['index'])
        labeled_probability = output_data/256
        recognitions = []
        for label, prob in zip(self.labels, labeled_probability[0]):
            recognitions.append(self.Recognition(label, prob))
        recognitions.sort(reverse=True)
        return recognitions[:min(ImageClassifier.MAX_SIZE, len(recognitions))]

    def load_image(self, image):
        crop_size = min(image.shape[0], image.shape[1])
        image = image[:crop_size, :crop_size]
        image = cv2.resize(image, (self.image_resize_x, self.image_resize_y))
        return image


    class Recognition:
        def __init__(self, name, confidence):
            self.name = name
            self.confidence = confidence

        def __str__(self):
            return f"Recognition{{name='{self.name}', confidence={self.confidence}}}"

        def __lt__(self, other):
            return self.confidence < other.confidence



model_path = "condition_model.tflite"
labels_path = "condition_labels.txt"
classifier = ImageClassifier(model_path, labels_path)
image_path = "images.jpg" 
recognitions = classifier.recognize_image(image_path)
if recognitions:
    for recognition in recognitions:
        print(recognition)
else:
    print("No recognitions found.")
```
